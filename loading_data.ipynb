**Loading the files with all the song and artist metadata** 

Got the data from this page: http://millionsongdataset.com/pages/find-song-specific-name-or-feature/

link: http://millionsongdataset.com/sites/default/files/AdditionalFiles/msd_summary_file.h5

```python
filename = 'msd_summary_file.h5'
f = h5py.File(filename, 'r')
f.keys()
```

Looking at musicbrainz

```python
pd.DataFrame(f['musicbrainz']['songs'][:])
```

*Nothing helpful there*

Looking at the metadata

```python
metadata = f['metadata']
metadata.keys()
```

```python
metadata_songs = pd.DataFrame(metadata['songs'][:])
metadata_songs = metadata_songs.convert_dtypes()
metadata_songs.head()
```

```python
metadata_songs.columns
```

```python
# Converting bytes to python obects
for col, d_type in metadata_songs.dtypes.to_dict().items():
    if d_type == object:
        print('Encoding column', col)
        try:
            metadata_songs[col] = metadata_songs[col].apply(lambda x: x.decode('ASCII'))
        except UnicodeDecodeError:
            metadata_songs[col] = metadata_songs[col].apply(lambda x: x.decode('UTF-8'))
```

```python
metadata_songs.head()
```

Joining song IDs to user data

```python
pd.merge(user_triplets, metadata_songs, left_on='song', right_on='song_id')
```

song and song_id are in fact the same attribute. It also looks to be an almost a perfect 1-to-1 match when joining both the user listening info and song metadata. 

**Looking at the analysis**

```python
d_set = f['analysis']
d_set.keys()
```

```python
million_song_df = pd.DataFrame(d_set['songs'][:])
million_song_df.head()
```

```python
million_song_df.columns
```

```python
# Converting bytes to python obects
for col, d_type in million_song_df.dtypes.to_dict().items():
    if d_type == object:
        print('Encoding column', col)
        try:
            million_song_df[col] = million_song_df[col].apply(lambda x: x.decode('ASCII'))
        except UnicodeDecodeError:
            million_song_df[col] = million_song_df[col].apply(lambda x: x.decode('UTF-8'))
million_song_df.head()
```

Joining million_song_df and all_tracks

```python
song_info = pd.merge(million_song_df, all_track, left_on='track_id', right_on='track_id')
song_info
```

```python
#Save song info
song_info.to_csv('./song_info.csv', index=False)
```

*Joining with metadata_songs*

joining on song_id from both datasets

```python
song_and_artist_info = pd.merge(song_info, metadata_songs, left_on='song_id', right_on='song_id')
song_and_artist_info.head()
```

```python
song_and_artist_info.columns
```

```python
# Saving he dataset
song_and_artist_info.to_csv('./song_and_artist_metadata.csv', index=False)
```
