```python
import pandas as pd
```

```python
df = pd.read_csv("./data/train_triplets_filtered.csv")
df
```

```python
def play_count_to_rating(play_count):
    if play_count <= 1:
        return 1
    if play_count <= 4:
        return 2
    if play_count <= 8:
        return 3
    if play_count <= 16:
        return 4
    return 5


df["rating"] = df.play_count.apply(play_count_to_rating)
df
```

```python
df["rating"].value_counts()
```

```python
df.to_csv("./data/train_triplets_filtered_ratings.csv", index=False)
```
