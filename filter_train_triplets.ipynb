```python
import pandas as pd
```

```python
df = pd.read_csv("./data/train_triplets.csv")
df
```

```python
prev_len = 0

while prev_len != len(df):
    prev_len = len(df)

    # Filter songs with >= 200 users
    counts = df.groupby(["song_id"])["user_id"].count()
    songs = counts.index[counts >= 200]
    df = df[df["song_id"].isin(songs)]

    # Filter users with >= 20 songs
    counts = df.groupby(["user_id"])["song_id"].count()
    users = counts.index[counts >= 20]
    df = df[df["user_id"].isin(users)]

    print(len(df))

df.reset_index(inplace=True, drop=True)
df
```

```python
df.to_csv("./data/train_triplets_filtered.csv", index=False)
```
